### Normalizing Data

```python
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
ndf = rdf.drop(['Sno','User_ID','Book_Title'],axis=1)
ndf.head()
```

```python
ndf = norm_func(ndf)
ndf.head()
```

```python
rdf = rdf.drop(['Book_Rating'],axis=1)
rdf.head()
```

### Concat Normalized Ratings with our dataframe

```python
tdf = pd.concat([rdf,ndf],axis=1)
tdf.head()
```

```python
userdf = tdf.pivot_table(index='User_ID',columns='Book_Title',values='Book_Rating').reset_index(drop=True)
```

```python
userdf.index= rdf.User_ID.unique()
```

```python
userdf
```

### Impute NaN with 0 values

```python
userdf.fillna(0,inplace=True)
```

```python
userdf
```

### Cosine method of measuring Similarity

```python
user_similarity = 1 - pairwise_distances(userdf.values,metric='cosine')
user_similarity
```

```python
user_similarity_df = pd.DataFrame(user_similarity)
user_similarity_df
```

```python
user_similarity_df.index = rdf.User_ID.unique()
user_similarity_df.columns = rdf.User_ID.unique()
user_similarity_df.iloc[0:5,0:5]
```

### Fill diagonal values with 0 as they will have similar values

```python
np.fill_diagonal(user_similarity,0)
user_similarity_df.iloc[0:5,0:5]
```

### Most Similar Users

## Fetching Rating for Fahrenheit 451

```python
Fahrenheit_rating = userdf['Fahrenheit 451']
Fahrenheit_rating
```

```python
Similar_to_Fahrenheit = userdf.corrwith(Fahrenheit_rating)
```

```python
Similar_to_Fahrenheit.head(20)
```

```python
corr_Fahrenheit = pd.DataFrame(Similar_to_Fahrenheit,columns=['Correlation'])
corr_Fahrenheit.dropna(inplace=True)
corr_Fahrenheit.head()
```

```python
corr_Fahrenheit[corr_Fahrenheit['Correlation']>-0.5].sort_values(by='Correlation',ascending=False).head(10)
```
